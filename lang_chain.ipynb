__參數設置說明__
1. chunk_size（塊大小）
建議設置：

文本分析：如果你需要對文本進行自然語言處理分析（例如情感分析、主題建模），較小的 chunk_size 可能會更合適（例如 200-500 字符）。

信息檢索：如果你需要從文本中檢索信息，適中的 chunk_size 可能會更合適（例如 500-1000 字符）。

上下文保留：如果需要保留更多上下文信息，較大的 chunk_size 可能會更合適（例如 1000-2000 字符）。

2. chunk_overlap（塊重疊）
建議設置：

連貫性：如果需要保留上下文連貫性，可以增加 chunk_overlap（例如設置為 chunk_size 的 10-20%）。

信息重複：如果希望避免信息重複，應減小 chunk_overlap，甚至設置為 0。

3. length_function（長度函數）
建議設置：

字符數：對於大多數文本處理任務，使用字符數（len 函數）通常是合適的。

單詞數：如果你更關心文本的單詞數，可以設置一個計算單詞數的函數。
